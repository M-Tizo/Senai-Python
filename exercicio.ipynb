```python
clientes = {
    123131312:{
        "saldo": 1000.00
    },
    414242:{
        "saldo":1000.00
    }
}


def validacao():
    while True:
        n_conta = input("Digite o numero da sua conta: ")
        global cpf
        cpf = int(input("Digite o seu cpf: "))
        if cpf in clientes:
            break
        else:
            print("Conta não cadastrada!")

def saque():
    while True:
        saque = float(input("Qual será o valor do saque? "))
        senha = int(input("Digite a sua senha: "))
        if senha != cpf:
            print("Senha incorreta, tente novamente!")
        
        else:
            break
    while True:
        saldo_depois = clientes[cpf]["saldo"] - saque
        if saldo_depois < 0:
            print("Você não pode sacar mais doque o disponível! Tente novamente!")
            saque = float(input("Qual será o novo valor de saque? "))
        
        else:
            clientes[cpf]["saldo"] = saldo_depois
            print(f"O saldo da sua conta após o saque será de R${clientes[cpf]["saldo"]}")
            break

def deposito():
    while True:
        deposito = float(input("Qual será o valor do deposito? "))
        senha = int(input("Digite a sua senha: "))
        if senha != cpf:
            print("Senha incorreta, tente novamente!")
        
        else:
            break
    while True:
        saldo_depois = clientes[cpf]["saldo"] + deposito
        clientes[cpf]["saldo"] = saldo_depois
        print(f"O saldo da sua conta após o deposito será de R${clientes[cpf]["saldo"]}")
        break

print("Bem-vindo ao sistema bancario!")
validacao()

while True:

    print(f"Seu saldo atual é de {clientes[cpf]}")
    opc = int(input("""
        1-Saque
        2-Deposito
        3-Sair
        Selecione uma das opções acima: """))
    if opc == 1:
        saque()
    elif opc == 2:
        deposito()
    elif opc == 3:
        print("Saindo do sistema....")
        break




        
        
```
